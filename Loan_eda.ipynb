```python
df_corr=df.corr()
```

```python
sns.boxplot(x='not.fully.paid',y='int.rate',data=df)
plt.title("Loan repayment vs interest rate")
plt.xlabel('not.fully.paid')
plt.ylabel('Interest rate')
```

Interest rate is highly positively correlated with Loan Repayment as higher the Interest rate will be more will be the chances of the loan getting  unpaid.

```python
sns.boxplot(x='not.fully.paid',y='dti',data=df)
plt.title("Loan repayment vs Debt to income ratio")
plt.xlabel('not.fully.paid')
plt.ylabel('Debt-to-income Ratio')
```

Debt to income ratio is positively correlated with Loan Repayment as higher the ratio more will be the risk of default.


```python
sns.boxplot(x='not.fully.paid',y='fico',data=df)
plt.title("Loan repayment vs fico score")
plt.xlabel('not.fully.paid')
plt.ylabel('fico score')
```

```python
sns.boxplot(x='not.fully.paid',y='revol.util',data=df)
plt.title("Loan repayment vs Revolving Utilization")
plt.xlabel('not.fully.paid')
plt.ylabel('revolving utilization')
```

```python
sns.boxplot(x='not.fully.paid',y='days.with.cr.line',data=df)
plt.title("Loan repayment vs Length of credit Line")
plt.xlabel('not.fully.paid')
plt.ylabel('Length of credit line')
```

Higher the Credit lines will be lesser will be the risk of default

```python
sns.boxplot(x='not.fully.paid',y='inq.last.6mths',data=df)
plt.title("Loan repayment vs Inquiries in last 6 months")
plt.xlabel('not.fully.paid')
plt.ylabel('Inquiries in last 6 months')
```

Higher number of credit inquiries in last 6 months may show positive correlation with default risk

```python
sns.barplot(x='not.fully.paid',y='purpose_small_business',data=df)
plt.title("Loan repayment vs purpose")
plt.xlabel('not.fully.paid')
plt.ylabel('Purpose(For Small Business)')
```

From above graph, most of the loan taken for starting small businesses are most likely getting unpaid.


## Correlations Influencing Default risk
1.Interest rate(int.rate)

Positive Correlation: int.rate seems to be positively correlated with not.fully.paid, which suggests that loans with higher interest rates are more likely to go unpaid. This makes sense, as higher interest payments increase the borrower's financial burden, potentially leading to default.

2.FICO(fico)

Negative Correlation: fico appears to have a negative correlation with not.fully.paid, implying that individuals with higher FICO scores (better credit ratings) are less likely to default.

3.Debt-to-income ratio(dti)

Positive Correlation: The debt-to-income ratio (dti) is likely positively correlated with not.fully.paid, indicating that borrowers with higher debt relative to their income may be more at risk of defaulting.

4.Revolving Balance (revol.bal) and Revolving Utilization (revol.util):

These features, particularly revol.util (revolving utilization rate), might show positive correlations with not.fully.paid. Higher values can indicate that a borrower is using a large portion of their available credit, which could signal financial stress.

5.Number of Inquiries in the Last 6 Months (inq.last.6mths)

Positive Correlation:Higher number of credit inquiries in last 6 months may show positive correlation with default risk

6.Purpose of loan

 Loans taken for purposes like small_business may have a stronger positive correlation with not.fully.paid, indicating a higher likelihood of default for small business loans.

##Summary
High Interest rate,Debt-to-income ratio and Lower FICO and some loan purpose can be the top indicators of Default risk.
